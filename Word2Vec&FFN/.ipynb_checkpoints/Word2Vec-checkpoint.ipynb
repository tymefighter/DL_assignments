### Import the Libraries

```python
import nltk
import math
import random
import numpy as np
import matplotlib.pyplot as plt
from math import floor
from sklearn.datasets import fetch_20newsgroups
from gensim.models import Word2Vec
from sklearn.manifold import TSNE
```

### Get the data

```python
newsgroups_train = fetch_20newsgroups(subset = 'all')
```

### Get the stop words

```python
nltk.download('stopwords')
stop_words = set(nltk.corpus.stopwords.words('english'))
```

### Tokenizer: Splits the sentences into tokens

```python
tokenizer = nltk.tokenize.RegexpTokenizer(r'\w+')
```

### Process the data
- Remove stop words from the data
- Place each sentence into a seperate list, and append each such small list to a larger list of sentences

```python
data = []


for line in newsgroups_train.data:
    
    line_arr = tokenizer.tokenize(line.lower())
    tokens = []
    
    for word in line_arr:
        if word not in stop_words:
            tokens.append(word)
            
    data.append(tokens)
```

```python
print("Number of sentences for training: " + str(len(data)))
```

# a) Word2Vec using gensim Word2Vec

### Train the model

```python
model = Word2Vec(data, size = 100, min_count = 10)
```

### Get vocab size

```python
print("Vocabulary size: " + str(len(model.wv.vocab)))
```

### Get the word vectors

```python
vectors = model.wv # Get word vectors
```

### Get the 10 most similar words to car along with their similarities 

```python
vectors.similar_by_word('car') # Get words similar to 'car'
```

### Get the 5 most similar words to the following vector

```python
vectors.similar_by_vector(vectors['girl'] + vectors['father'] - vectors['boy'], topn = 5)
```

### Get the 5 most similar words to the following vector

```python
vectors.similar_by_vector(vectors['sports'] - vectors['bat'] + vectors['ball'], topn = 5)
```

### Reduce the dimensionality to visualize the words in 2 dimensions

```python
tsne = TSNE(n_components = 2) # Get TSNE object
```

### Get 20 words most similar to the following words and plot them

```python
words = ['baseball', 'software', 'police', 'government', 'circuit', 'car']
colors = ['magenta', 'blue', 'cyan', 'green', 'orange', 'red']

sim_words = []
sim_vecs = []


for i in range(len(words)):
    
    word = words[i]
    similar_word_tuples = vectors.similar_by_word(word, topn = 20)
    
    for word_tup in similar_word_tuples:
        
        sim_words.append(word_tup[0])
        sim_vecs.append(vectors[word_tup[0]])
    

sim_vec_lower = tsne.fit_transform(np.array(sim_vecs))
```

```python
fig, ax = plt.subplots()
fig.set_figheight(10)
fig.set_figwidth(20)

for i in range(len(sim_words)):
    
    color = colors[i // 20]
    
    if i % 20 == 0:
        ax.scatter(sim_vec_lower[i][0], sim_vec_lower[i][1], c = color, label = words[i // 20])
    else:
        ax.scatter(sim_vec_lower[i][0], sim_vec_lower[i][1], c = color)
    ax.text(sim_vec_lower[i][0], sim_vec_lower[i][1], s = sim_words[i])

plt.legend()
plt.show()
```

# b) Implementing a classifier to classify a sentence to one among the 20 news groups

### Import TensorFlow to build a classifier

```python
import tensorflow as tf
from sklearn.metrics import confusion_matrix
```

### Training the model (word2vec) using the Skip - Gram method

```python
model_sg = Word2Vec(data, size = 100, min_count = 10, sg = 1)
```

```python
vectors_sg = model_sg.wv
```

### Training the model (word2vec) using the CBOW method

```python
model_cbow = Word2Vec(data, size = 100, min_count = 10, sg = 0)
```

```python
vectors_cbow = model_cbow.wv
```

### Getting labelled dataset

```python
print("Target names: ", newsgroups_train.target_names)
```

```python
train_data = data[ : floor(0.7 * len(data))]
train_target = newsgroups_train.target[ : floor(0.7 * len(data))]

val_data = data[floor(0.7 * len(data)) : floor(0.8 * len(data))]
val_target = newsgroups_train.target[floor(0.7 * len(data)) : floor(0.8 * len(data))]

test_data = data[floor(0.8 * len(data)) : ]
test_target = newsgroups_train.target[floor(0.8 * len(data)) : ]
```

```python
print("Train data: ", len(train_data), "Train target: ", len(train_target))
print("Val data: ", len(val_data), "Val target: ", len(val_target))
print("Test data: ", len(test_data), "Test target: ", len(test_target))
```

#### Function to plot the loss and accuracy curves

```python
def plot_arr(arr, xlabel, ylabel, title):
    
    plt.plot(arr)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
    plt.title(title)
    plt.show()
```

## Method 1: Averaging the word vectors in the sentences (Skip - Gram)

```python
X_train_sg = np.zeros((len(train_data), 100))

for i in range(len(train_data)):
    num = 0
    for word in train_data[i]:
        if word in vectors_sg:
            X_train_sg[i] += vectors_sg[word]
            num += 1
    
    X_train_sg[i] /= num


X_val_sg = np.zeros((len(val_data), 100))

for i in range(len(val_data)):
    num = 0
    for word in val_data[i]:
        if word in vectors_sg:
            X_val_sg[i] += vectors_sg[word]
            num += 1
    
    X_val_sg[i] /= num

X_test_sg = np.zeros((len(test_data), 100))

for i in range(len(test_data)):
    num = 0
    for word in test_data[i]:
        if word in vectors_sg:
            X_test_sg[i] += vectors_sg[word]
            num += 1
    
    X_test_sg[i] /= num
```

```python
Y_train = np.array(train_target)
Y_val = np.array(val_target)
Y_test = np.array(test_target)
```

### i) Model 1

- Skip - Gram vectors used to build document vector
- 1 hidden layer with 100 hidden units and relu activation function
- Loss: Sparse Categorical Crossentropy

```python
model11 = tf.keras.Sequential([
    tf.keras.layers.Dense(200, activation=tf.nn.relu),
    tf.keras.layers.Dense(20, activation=tf.nn.softmax)
])

model11.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['acc'])

hs = model11.fit(X_train_sg, Y_train, epochs = 100, verbose = 0)
```

```python
history = hs.history
```

```python
plot_arr(history['loss'], 'num iters', 'loss value', 'loss vs iters')
```

```python
plot_arr(history['acc'], 'num iters', 'accuracy', 'accuracy vs iters')
```

```python
plot_arr((1 - np.array(history['acc'])) * 100, 'num iters', 'classification error', 'ce vs iters')
```

```python
validation_loss, validation_acc = model11.evaluate(X_val_sg, Y_val, verbose = 0)
print("Validation Loss: ", validation_loss)
print("Validation Acc: ", validation_acc)
```

### ii) Model 2

- Skip - Gram vectors used to build document vector
- 2 hidden layers with 100 hidden units and relu activation function
- Loss: Sparse Categorical Crossentropy

```python
model12 = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation=tf.nn.relu),
    tf.keras.layers.Dense(100, activation=tf.nn.relu),
    tf.keras.layers.Dense(20, activation=tf.nn.softmax)
])

model12.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['acc'])

hs = model12.fit(X_train_sg, Y_train, epochs = 100, verbose = 0)
```

```python
history = hs.history
```

```python
plot_arr(history['loss'], 'num iters', 'loss value', 'loss vs iters')
```

```python
plot_arr(history['acc'], 'num iters', 'accuracy', 'accuracy vs iters')
```

```python
plot_arr((1 - np.array(history['acc'])) * 100, 'num iters', 'classification error', 'ce vs iters')
```

```python
validation_loss, validation_acc = model12.evaluate(X_val_sg, Y_val, verbose = 0)
print("Validation Loss: ", validation_loss)
print("Validation Acc: ", validation_acc)
```

## Method 2: Averaging the word vectors in the sentences (CBOW)

```python
X_train_cbow = np.zeros((len(train_data), 100))

for i in range(len(train_data)):
    num = 0
    for word in train_data[i]:
        if word in vectors_cbow:
            X_train_cbow[i] += vectors_cbow[word]
            num += 1
    
    X_train_cbow[i] /= num


X_val_cbow = np.zeros((len(val_data), 100))

for i in range(len(val_data)):
    num = 0
    for word in val_data[i]:
        if word in vectors_cbow:
            X_val_cbow[i] += vectors_cbow[word]
            num += 1
    
    X_val_cbow[i] /= num

X_test_cbow = np.zeros((len(test_data), 100))

for i in range(len(test_data)):
    num = 0
    for word in test_data[i]:
        if word in vectors_cbow:
            X_test_cbow[i] += vectors_cbow[word]
            num += 1
    
    X_test_cbow[i] /= num
```

### i) Model 1

- CBOW vectors used to build document vector
- 1 hidden layer with 100 hidden units and relu activation function
- Loss: Sparse Categorical Crossentropy

```python
model21 = tf.keras.Sequential([
    tf.keras.layers.Dense(200, activation=tf.nn.relu),
    tf.keras.layers.Dense(20, activation=tf.nn.softmax)
])

model21.compile(optimizer = 'adam', loss = 'sparse_categorical_crossentropy', metrics = ['acc'])

hs = model21.fit(X_train_cbow, Y_train, epochs = 100, verbose = 0)
```

```python
history = hs.history
```

```python
plot_arr(history['loss'], 'num iters', 'loss value', 'loss vs iters')
```

```python
plot_arr(history['acc'], 'num iters', 'accuracy', 'accuracy vs iters')
```

```python
validation_loss, validation_acc = model21.evaluate(X_val_cbow, Y_val, verbose = 0)
print("Validation Loss: ", validation_loss)
print("Validation Acc: ", validation_acc)
```

### i) Model 2

#### Getting TP, TN, FP and FN for the Training Data

```python

cnf_matrix_train = confusion_matrix(np.argmax(model11.predict(X_train_sg), axis = 1), Y_train)

TP = np.diag(cnf_matrix_train)
FP = cnf_matrix_train.sum(axis = 0) - np.diag(cnf_matrix_train)
FN = cnf_matrix_train.sum(axis = 1) - np.diag(cnf_matrix_train)
TN = cnf_matrix_train.sum() - (TP + FP + FN)

print("For Training Data: -")
for cl in range(20):
    
    print("Class %d: TP = %d, TN = %d, FP = %d, FN = %d" %(cl + 1, TP[cl], TN[cl], FP[cl], FN[cl]))
```

#### Getting Test Loss and Test Accuracy

```python
test_loss, test_acc = model11.evaluate(X_test_sg, Y_test, verbose = 0)
print("Test Loss: ", test_loss)
print("Test Acc: ", test_acc)
```

#### Getting TP, TN, FP and FN for the Test Data

```python
cnf_matrix_test = confusion_matrix(np.argmax(model11.predict(X_test_sg), axis = 1), Y_test)

TP = np.diag(cnf_matrix_test)
FP = cnf_matrix_test.sum(axis = 0) - np.diag(cnf_matrix_test)
FN = cnf_matrix_test.sum(axis = 1) - np.diag(cnf_matrix_test)
TN = cnf_matrix_test.sum() - (TP + FP + FN)

print("For Test Data: -")
for cl in range(20):
    
    print("Class %d: TP = %d, TN = %d, FP = %d, FN = %d" %(cl + 1, TP[cl], TN[cl], FP[cl], FN[cl]))
```
